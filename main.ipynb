```python
def get_nulls_columns(df:pd.DataFrame, column : str = '') -> pd.DataFrame:
    if column == '':
        return df.loc[df.isnull().any(axis=1)]
    else:
        return df[df[column].isnull()]


def get_rid_of_nulls(df: pd.DataFrame, approach: str) -> pd.DataFrame:
    df = df.copy()
    approaches = {
        'drop': lambda dataset: dataset.dropna(),
        'numeric-mean': lambda dataset: dataset[dataset.select_dtypes(include=['int64', 'float64']).columns].fillna(dataset.mean()),
        'statistical-avg': lambda dataset: get_statistical_avg(dataset)
    }
    operation = None
    try: 
        operation = approaches[approach]
    except KeyError:
        print(f"There is no option '{approach}'")
        print(f"Try using one of the following: {', '.join(approaches.keys())}")
        print("For better understanding look in the documentaion")

        raise KeyError('Approach not found')
    except TypeError as e:
        print(f"Error: {e}. Make sure you are passing a DataFrame to the function.")
    if operation:
        if 'numeric' in approach:
            numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
            df[numeric_cols] = operation(df[numeric_cols])
        else:
            df = operation(df)

    return df

```
